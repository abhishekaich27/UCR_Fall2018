## <span style="color:blue">Initializations:</span>

$\bullet$ Import required packages:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import Markdown, display
def printmd(string):
    display(Markdown(string))
    
# Set figure size and resolution
plt.rcParams['figure.figsize'] = [4, 3]
plt.rcParams['figure.dpi']= 200
```

$\bullet$ Import dataset file _five-stocks.csv_ :

```python
comp_SM = pd.read_csv("five_stocks.csv")
SM = comp_SM.values
```

$\bullet$ Mean, $~\large\mu_k = \sum\limits_{i=(k-1)T+1}^{kT}r_i$

```python
def mean_func(col, k, T):
    
    start = ((k - 1) * T) - 1
    end = k * T
    sum_col = 0
    
    for i in range(start, end):
        sum_col = sum_col + col[i]
    
    return sum_col/T
```

$\bullet$ Empirical Auto-correlation, $~\large R(\tau) = \dfrac{1}{N-\tau}\sum\limits_{i=1}^{N-\tau}r_ir_{i+\tau}$

```python
def corr_func(X, Y, tau):

    N = np.size(X,0)
    R_tau = 0

    for ind in range(N - tau):
        R_tau = R_tau + (X[ind] * Y[ind + tau])

    R_tau = R_tau / (N - tau)

    return R_tau
```

$\bullet$ Correlation Coefficient, $~\large \rho(i,k) = \dfrac{1}{N}\sum\limits_{i=1}^{N}\dfrac{(r_{j,i} - \mu_j)(r_{k,i} - \mu_k)}{\sigma_j\sigma_k}$

```python
def corr_coeff(X, Y):
    
    X = X - X.mean(1)[:,None]
    Y = Y - Y.mean(1)[:,None]
    var_X = (X**2).sum(1);
    var_Y = (Y**2).sum(1);

    return np.dot(X, Y.T)/np.sqrt(np.dot(var_X[:,None], var_Y[None]))
```

$\bullet$ Estimates the maximum correlation among the columns:

```python
def max_corr(A):
    
    A = A - np.eye(5)
    i, j = np.unravel_index(np.argmax(A), A.shape)
    
    comp_name = list(comp_SM.columns.values)
    
    return comp_name[i], comp_name[j]
```

$\bullet$ (Exponentiated) Sum-process, $\large q_i = \dfrac{p_i}{p_0} = \exp(\sum\limits_{j=1}^i r_j)$

```python
def normalized_price(SM):
    
    comp_name = list(comp_SM.columns.values)
    
    q = np.eye(504,5)
    for i in range(5):
        for j in range(504):
            r = comp_SM[comp_name[i]].values 
            q[j,i] = np.exp(np.sum(r[:j+1]))
            
    return q
```

## <span style="color:blue">1. Calculate the empirical autocorrelation $R(\tau)$ (which is $R_{\text{GOOGL}}(\tau)$ and autocovariance $C(\tau)$ from day-offset $\tau = 0$ to $\tau = 125$ and plot it.</span> 

```python
R_tau  = np.zeros(251)
C_tau  = np.zeros_like(R_tau)

GOOGL = SM[:,2]

for tau in range(126):
    R_tau[tau+125] = corr_func(GOOGL, GOOGL, tau)
    R_tau[125-tau] = R_tau[tau+125]
    C_tau[tau+125] = corr_func(GOOGL - np.mean(GOOGL), GOOGL - np.mean(GOOGL), tau)
    C_tau[125-tau] = C_tau[tau+125]

t = np.arange(-125,126)    

# Plot auto-correlation
plt.plot(t, R_tau, color = 'black')
plt.title("Auto-correlation for Google", color='r')
plt.xlabel("$\\tau$")
plt.ylabel("$R_{GOOGL}(\\tau)$")
plt.savefig('R_tau_Google.pdf', dpi=300, bbox_inches = 'tight')
plt.grid(True)
plt.show()

# Plot auto-covariance
plt.plot(t, C_tau, color = 'g')
plt.title("Auto-covariance for Google", color='r')
plt.xlabel("$\\tau$")
plt.ylabel("$C_{GOOGL}(\\tau)$")
plt.savefig('C_tau_Google.pdf',dpi=300, bbox_inches = 'tight')
plt.grid(True)
plt.show()
```

<b> Q. Brief comment on $R(0)$ (or $C(0)$) compared to other values. </b><br>
Ans: The stocks of maximum correlation at $\tau = 0$, and which is expectedly more than value of $R(\tau)$ at $\tau\neq 0$. The same goes for $C(\tau)$ as well. Consequently we can say that today’s return will not be correlated with future returns.

## <span style="color:blue">2. Take the (discrete) Fourier transform of the autocorrelation $C(\tau)$ and plot it from −125 to 125. Brief comment on the spectrum.</span> 

```python
PS = np.fft.fft(C_tau)

plt.plot(t, PS, color = 'g')
plt.title("Power Spectrum from $R_{GOOGL}(\\tau)$", color='r')
plt.savefig('Power_Spectrum_Google.pdf',dpi=300, bbox_inches = 'tight')
plt.grid(True)
plt.show()
```

<b> Q. Brief comment on the spectrum. </b><br>
Ans: The spectrum is almost constant around amplitude 0. As $R(\tau)$ is uncorrelated in time, it's power spectrum is uniformly spread across all allowed frequencies.

## <span style="color:blue">3. Create a bar plot where $x$ axis runs from 1 to 8 and $y$ axis is the average return.</span> 

```python
GOOGL = SM[:,2]

k_values = np.arange(1, 9)
mean_k = []
for k in k_values:
    mean = mean_func(GOOGL, k, 63)
    mean_k.append(mean)

plt.bar(k_values, mean_k, color = 'g')
plt.title("Mean for Different Quarters for Google", color='r')
plt.ylabel("Average return")
plt.savefig('quarter_mean.pdf',dpi=300, bbox_inches = 'tight')
plt.grid(True)
plt.show()
```

<b> Q. Is this stationary? </b><br>
Ans: No. The mean is not steady for all quarters.

## <span style="color:blue">4(a). Calculate the empirical cross-correlation between Amazon and Google i.e. find empirical $R_{XY}(\tau)$ where $X$ = Google and $Y$ = Amazon by modifying (2).</span>

```python
GOOGL = SM[:,2]
AMZN = SM[:,0]

R_XY_tau  = np.zeros(251)
ti = np.arange(-125,126)

for tau in range(126):
    R_XY_tau[tau + 125] = corr_func(GOOGL, AMZN, tau)
    R_XY_tau[125 - tau] = R_XY_tau[tau + 125]

plt.plot(ti, R_XY_tau, color = 'r')
plt.title("Cross-correlation b/w Amazon and Google", color='r')
plt.xlabel("$\\tau$")
plt.ylabel("$C_{GOOGL,AMZN}(\\tau)$")
plt.savefig('corr_Google_Amazon.pdf',dpi=300, bbox_inches = 'tight')
plt.grid(True)
plt.show()
```

<b> Q. Brief comment on the plot. </b><br>
Ans: Again, the most correlation between the stock prices of Amazon and Google can be seen at $\tau = 0$.
